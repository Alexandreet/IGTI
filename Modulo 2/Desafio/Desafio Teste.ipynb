```python
for jogadores in jogador.itertuples():
    nome_jogador = jogadores.nome_jogador
    genero = jogadores.genero
    data_nascimento = str(jogadores.data_nascimento)
    cod_jogo = jogadores.cod_jogo
    cod_pais = jogadores.cod_pais
    num_vitorias = jogadores.num_vitorias
    num_derrotas = jogadores.num_derrotas
    total_partidas = jogadores.total_partidas
    try:
        query = f"""insert into jogador (nome_jogador, genero, data_nascimento, cod_jogo, cod_pais, num_vitorias, num_derrotas, total_partidas) 
                     values ('{nome_jogador}', '{genero}',  '{data_nascimento}', '{cod_jogo}', '{cod_pais}', '{num_vitorias}', '{num_derrotas}', '{total_partidas}')"""      
        conn.execute(query)      
        print(f"Registro inserido com sucesso: {nome_jogador}")
    except Exception as e:        
        print(f"Não foi possível inserir o regristro {nome_jogador}. O erro foi encontrado foi: {e}")    
```
