#### Neste exercício vamos pegar vários arquivos separados de vendas, e unifica-los em uma única base de dados

```python
#Importando biblitocas para a excecução do arquivo
import pandas as pd
import os
import MySQLdb
import sqlalchemy
```

```python
#criando o caminho até o diretório onde se encontram as tabelas
DIRETORIO = "/Users/macbook/Documents/Projects Github/IGTI/Modulo 2/Material_complementar/Vendas veiculos/dataset/vendas_veiculos/"
```

```python
lista_df = [] #criando uma lista vazia
lista_path = os.listdir(DIRETORIO)
```

```python
#encontrado um erro nos arquivos de mac onde se lê os arquivos ocultos.DS_Store, por isso usando .remove()
lista_path.remove('.DS_Store')
lista_path[:]
```

```python
#Criando uma lista vazia
lista_df = []
#criando loop for onde vamos armazenar no dataframe 'df' os arquivos dentro do DIRETORIO
#criamos uma variavel local para armazenar a função os.path.join para quando for determinar o parametro de caminho
#do pd.read_csv termos um caminho que pode ser repetido varias vezes dentro do loop for para cada 'arquivo'
#armazenamos os conteudos do loop for dentro da variavel lista_df e depois fazemos um pd.concat no df_vendas
#para transformas lista_df em um dataframe
for arquivo in lista_path:
    local_arquivo = os.path.join(DIRETORIO, arquivo)
    df = pd.read_csv(local_arquivo, sep = ';', encoding = 'latin-1')
    lista_df.append(df)
df_vendas = pd.concat(lista_df)
```

```python
df_vendas[:10]
```

## Exportando o Data Frame

```python
#exportando o novo daframe criado anteriormente com a sep em ';' e encoding em 'latin-1' SEM o index
#pois o index está bagunçado por conta de ser uma junção de dataframes
df_vendas.to_csv(DIRETORIO + 'vendas_por_data.csv', sep = ';', encoding = 'latin-1', index = False)
```

```python
#Criando as tabelas individuais apartir do df_vendas eliminando duplicatas
#sendo que fabricantes por só ter um item só precisa ser uma lista
df_pessoas = df_vendas[['cod_pessoa', 'nome', 'genero', 'data_nascimento']].drop_duplicates()
df_modelos = df_vendas[['modelo', 'fabricante']].drop_duplicates()
fabricantes = list(df_vendas.fabricante.unique())
```

```python
#Nova função itertuples vai fazer várias iterações criando tuplas  chave e valor dos dados da tabela
#definimos cada valor relativo ao valor dentro da base de dados, com o caminho "nome_tabela.nome_campo_da_tabela"
#tudo isso dentro de laço de repetição for
for pessoa in df_pessoas.itertuples():
    cod_pessoa = pessoa.cod_pessoa
    nome = pessoa.nome
    genero = pessoa.genero
    data_nascimento = str(pessoa.data_nascimento)
    try:
        query = f"""insert into pessoa (cod_pessoa, nome, genero, data_nascimento )
                        values('{cod_pessoa}', '{nome}', '{genero}', '{data_nascimento}')"""
        conn.execute(query)
        print(f"Registo foi inserido com sucesso: {nome}")
    except Exception as e:
        print(f"Não foi possivel inserir o regristro {nome}. Erro encontrado foi: {e}")
```

```python
#Repitimos o mesmo processo só que para uma lista
#Lembrando que na hora de criar o database no mysql devemos adicionar a função AUTO_INCREMENT
for fabricante in fabricantes:
    try:
        query = f"""insert into fabricante (nome_fabricante)
                    values('{fabricante}')"""
        conn.execute(query)
        print(f"Registro inserido com secesso: {fabricante}")
    except Exeption as e:
        print(f"Não foi possivel inserir o registro {fabricante}. O erro encontrado foi: {e}")
```

```python
#Vamos inserir dentro da tabela vendas_veiculos dentro do sql o codigo do fabricante INT que não possuimos
df_fabricante = pd.read_sql('fabricante', con=conn)
```

```python
df_fabricante
```

```python
#Vamos criar um dicionario com chaves sendo nome_fabricante e cod_fabricante sendo os valores
dict_fabricante = df_fabricante.set_index(df_fabricante.nome_fabricante)['cod_fabricante'].to_dict()
dict_fabricante
```

```python
#Usando a função map vamos encontrar e substituir os valores do fabricante dentro da tabela df_modelos
#para atender a dependencia que temos na base de dados com o codigo do fabricante
df_modelos.fabricante = df_modelos.fabricante.map(dict_fabricante)
df_modelos
```

```python
df_modelos = df_modelos.rename(columns = {'fabricante' : 'cod_fabricante'})
df_modelos
```

```python
for modelo in df_modelos.itertuples():
    nome_modelo = modelo.modelo
    cod_fabricante = modelo.cod_fabricante
    try:
        query = f"""insert into modelo_veiculo(descricao_modelo, cod_fabricante)
                        values('{nome_modelo}','{cod_fabricante}')"""
        conn.execute(query)
        print(f"Registro inserido com sucesso: {nome_modelo}")
    except Exception as e:
        print(f"Não foi possivel inserir o registro: {nome_modelo}. Erro:{e}")
```
