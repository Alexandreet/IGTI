#### Neste exercício vamos pegar vários arquivos separados de vendas, e unifica-los em uma única base de dados

```python
#Importando Biblioteca Pandas
import pandas as pd
#Importanto bibliotecas e funções do sistema operação
import os
```

```python
#Criando uma variável que será utilizada para substituir o caminho até os arquivos
diretorio = "/Users/macbook/Documents/Projects Github/IGTI/Modulo 2/Material_complementar/Vendas veiculos/dataset/vendas_veiculos/"
```

```python
#Criando uma lista para armazenar os arquivos das bases de dados
lista_df = []
#Criar uma variável usando a função do OS listdir, que optminizará ainda mais o código
lista_path = os.listdir(diretorio)
```

```python
#Verificação de que o path está no diretório correto
lista_path[:5]
```

```python
for arquivo in lista_path: #criando um loop for para repetir as ações seguintes
    local_arquivo = os.path.join(diretorio, arquivo) #Usando a função do os Path.join para que sejam localizados os arquivos
    df = pd.read_csv(local_arquivo, sep = ';', encoding = 'latin-1') #criando um dataframe temporarios para adicionar todos os arquivos do csv com separação por ';' e enconding em latin-1
    lista_df.append(df) #usando a função appennd para incluir os valores do df temporario dentro da lista criada no começo
df_vendas = pd.concat(lista_df) #convertendo a lista em data frame com a função do pandas concat e salvando no novo data frame df_vendas
```

```python
#verificando o novo df 
df_vendas[:10]
```

```python
#Apresentado um erro com uma coluna a mais chadama shellclassinfo, removemos a coluna criando um novo dataframe
#sem ela através da função drop com o parametro columns
df_vendas_export = df_vendas.drop(columns=['[.ShellClassInfo]'])
```

```python
df_vendas_export
```

```python
#Usando a função to_csv podemos exportar data base tratada para um arquivo csv dentro do diretório já pré estabelecido
df_vendas_export.to_csv(diretorio + 'vendas_por_data.csv', sep = ';', encoding = 'latin-1', index = False)
```

```python
for pessoa in df_pessoas.itertuples():
    cod_pessoa = pessoa.cod_pessoa
    nome = pessoa.nome
    genero = pessoa.genero
    data_nascimento = str(pessoa.data_nascimento)
    try:
        query = f"""insert into pessoa (cod_pessoa, nome, genero, data_nascimento)
                    values('{cod_pessoa}' ,'{nome}', '{genero}', '{data_nascimento}')"""
        conn.execute(query)
        print(f"Registro inserido com sucesso: {nome}")
    except  Exception as e:
        print(f"Não foi possivel inserir o registro {nome}. O erro encontrado foi: {e}")
```

```python
for fabricante in fabricantes:
    try:
        query = f"""insert into fabricante (nome_fabricante) 
                     values ('{fabricante}')"""
        conn.execute(query)
        print(f"Registro inserido com sucesso: {fabricante}")
    except Exception as e:
        print(f"Não foi possível inserir o regristro {fabricante}. O erro foi encontrado foi: {e}")    
```
