## Coleta dos Grandes Mestres do Xadrez

```python
url = 'https://api.chess.com/pub/titled/GM'
```

```python
response = requests.get(url)
```

```python
response
```

```python
response.json()
```

```python
lista_jogadores = response.json()
```

```python
len(lista_jogadores['players'])
```

```python
def busca_usuario(usuario):
    url = f'https://api.chess.com/pub/player/{usuario}'
    response = requests.get(url)
    if response.ok:
        return response.json()
```

```python
%%time
busca_usuario(lista_jogadores['players'][0])
```

## Criando uma estrutura que guarda todos os dados

```python
lista_usuario = []
for usuario in lista_jogadores['players']:
    dados_usuario = busca_usuario(usuario)
    print(f'Coleta realisada com sucesso para o jogador: {usuario}')
    lista_usuario.append(dados_usuario)
```

```python
lista_usuario[:3]
```

```python
import pandas as pd
```

```python
df_jogadores = pd.DataFrame(lista_usuario)
df_jogadores[:3]
```

```python
df_jogadores.to_csv('jogadores_chess_com.csv', sep = ';')
```
