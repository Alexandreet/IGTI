```python
import tweepy as tw
import pandas as pd
```

```python
with open ("twitter_token.txt", "r") as file:
    consumer_key = file.readline().strip('\n')
    consumer_secret = file.readline().strip('\n')
    token = file.readline().strip('\n')
    access_token = file.readline().strip('\n')
    access_token_secret = file.readline().strip('\n')
```

```python
auth = tw.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tw.API(auth)

public_tweets = api.home_timeline()
for tweet in public_tweets:
    print(tweet.text)
```

## Criando um Data Frame para guardar os tweets

```python
tweets = tw.Cursor(api.search, q=busca_palavras).items(1500)
```

```python
lista_tweets = []
for tweet in tweets:
    dados_tweet = []
    dados_tweet.append(tweet.created_at)
    dados_tweet.append(tweet.user.name)
    dados_tweet.append(tweet.user.description)
    dados_tweet.append(tweet.user.location)
    dados_tweet.append(tweet.text)
    
    lista_tweets.append(dados_tweet)
```

```python
lista_tweets[:3]
```

```python
df_tweets = pd.DataFrame(lista_tweets, columns=['data','nome','descricao','localidade','post'])
```

```python
conn = MongoClient('localhost', 27017)
type(conn)
```

```python
#Vamos criar implicitamente um banco de dados chamado tweetsdb
db = conn.tweetsdb
type(db)
```

```python
#Vamos criar uma cole√ß√£o de documentos para armazenar no MongoDB
collection = db.tweets_coletados
type(collection)
```

### Transformando dataframe em dicion√°rio

```python
tweets_dict = df_tweets.to_dict('records')
tweets_dict[:3]
```

#### Alimentando a cole√ß√£o com os dados do dicion√°rio e colocando um "_id" automaticamente apartir do comando inserted_id caso o documento n√£o possua a chave _id

```python
for tweet in tweets_dict:
    post_id = collection.insert_one(tweet)
    post_id.inserted_id
```
